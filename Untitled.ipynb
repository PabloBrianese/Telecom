1. Calcular la facturación promedio, máxima, mínima de los últimos tres meses de los clientes 12-15-20.

```python
with sqlite3.connect('file:telecom.db?mode=ro', uri=True) as con:
    cur = con.cursor()

    query = con.execute("""
        SELECT
            mes,
            MIN(facturacion),
            AVG(facturacion),
            MAX(facturacion)
        FROM facturacion
        WHERE nro_cliente IN (12, 15, 20)
        GROUP BY mes
        ORDER BY mes;
    """)

    column_names = tuple(row[0] for row in query.description)
    rows = query.fetchall()

    pprint([column_names] + rows)
```

2. Crear una marca en caso de que haya cambiado de plan en alguno de los tres meses. (subilo al repositorio en un archivo .csv)

```python
with sqlite3.connect("telecom.db") as con:
    cur = con.cursor()

    cur.execute("""
        CREATE TABLE IF NOT EXISTS cambio_de_plan AS
        SELECT
            nro_cliente,
            CASE
                WHEN COUNT(DISTINCT Plan_mes) > 1 THEN TRUE
                ELSE FALSE
            END AS cambio_de_plan
        FROM cliente
        GROUP BY nro_cliente;
    """)

    con.commit()

with sqlite3.connect('file:telecom.db?mode=ro', uri=True) as con:
    cur = con.cursor()

    query = cur.execute("SELECT * FROM cambio_de_plan;")

    column_names = tuple(col[0] for col in query.description)
    rows = query.fetchall()

    pprint([column_names] + rows)
```

```python
with open('cambio_de_plan.csv', 'w', newline='') as file:
  wtr = csv.writer(file, delimiter=';')
  wtr.writerows([column_names] + rows)
```

4. Que provincia tiene la facturación promedio trimestral más alta


```python
with sqlite3.connect('file:telecom.db?mode=ro', uri=True) as con:
  cur = con.cursor()

  query = cur.execute("""
    SELECT
      AVG(facturacion) AS facturacion_promedio_trimestral,
      provincia
    FROM facturacion
    INNER JOIN provincia
    ON facturacion.nro_cliente = provincia.nro_cliente
    GROUP BY provincia
    ORDER BY facturacion_promedio_trimestral DESC
    LIMIT 1;
  """)

  column_names = tuple(row[0] for row in query.description)
  rows = query.fetchall()

  pprint([column_names] + rows)
```

5. Hay alguna relación entre la facturación y la duración de llamadas (hay alguna métrica estadística que lo sustente)

La información sobre la facturación se encuentra en la tabla `facturacion`, y la información sobre la duración de las llamadas se encuentra en la tabla `llamadas`. Ambas tablas están conectadas por la columna `nro_cliente`. Luego, debemos extraer la relación entre las columnas `facturacion.facturacion`, `llamadas.Duracion_Llamadas_minutos` y la columna `nro_cliente`. 

En la tabla `llamadas`, hay muchos registros distintos para cada cliente con un `nro_cliente` particular. Por eso tenemos que agregar los minutos de llamadas agrupados por cliente. Registramos el resultado en una nueva tabla.

```python
with sqlite3.connect("telecom.db") as con:
  cur = con.cursor()

  cur.execute("""
    CREATE TABLE IF NOT EXISTS duracion_llamadas_minutos_en_trimestre
    AS
    SELECT
      SUM(Duracion_Llamadas_minutos) AS duracion_llamadas_minutos_en_trimestre,
      nro_cliente
    FROM llamadas
    GROUP BY nro_cliente
    ORDER BY duracion_llamadas_minutos_en_trimestre;
  """)

  con.commit()
```

```python
with sqlite3.connect('file:telecom.db?mode=ro', uri=True) as con:
  cur = con.cursor()

  query = cur.execute("SELECT * FROM duracion_llamadas_minutos_en_trimestre;")

  column_names = tuple(row[0] for row in query.description)
  rows = query.fetchall()

  pprint([column_names] + rows)
```

Con la facturación `facturacion` sucede algo similar, está registrada por mes. Nosotros necesitamos la relación que hay entre la facturación y un cliente determinado. Por eso agregamos la facturación agrupando por número de cliente.

```python
with sqlite3.connect("telecom.db") as con:
  cur = con.cursor()

  cur.execute("""
    CREATE TABLE IF NOT EXISTS facturacion_en_trimestre
    AS
    SELECT
      SUM(facturacion) AS facturacion_en_trimestre,
      nro_cliente
    FROM facturacion
    GROUP BY nro_cliente
    ORDER BY facturacion_en_trimestre;
  """)

  con.commit()
```

```python
with sqlite3.connect('file:telecom.db?mode=ro', uri=True) as con:
  cur = con.cursor()

  query = cur.execute("SELECT * FROM facturacion_en_trimestre;")

  column_names = tuple(row[0] for row in query.description)
  rows = query.fetchall()

  pprint([column_names] + rows)
```

Ahora podemos juntar toda la información para estudiar la relación entre la facturación y la duración de las llamadas.

```python
with sqlite3.connect('file:telecom.db?mode=ro', uri=True) as con:
  cur = con.cursor()

  query = cur.execute("""
    SELECT
      facturacion_en_trimestre,
      duracion_llamadas_minutos_en_trimestre
    FROM facturacion_en_trimestre AS f
    INNER JOIN duracion_llamadas_minutos_en_trimestre AS l
    ON f.nro_cliente = l.nro_cliente
    ORDER BY facturacion_en_trimestre;
  """)

  column_names = tuple(row[0] for row in query.description)
  rows = query.fetchall()

  pprint([column_names] + rows)
```

Usamos herramientas de cómputo, estadísticas y de visualización para estudiar las distribuciones de probabilidad que describen las variables _facturación_ y _duración de la llamada_.

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
import seaborn as sns

facturaciones = tuple(row[0] for row in rows)
duraciones = tuple(row[1] for row in rows)
```

```python
plt.title('distribución conjunta')

plt.xlabel('duracion')
plt.ylabel('facturacion')

sns.scatterplot(x=duraciones, y=facturaciones)
```

```python
plt.title('distribución individual')
plt.xlabel('facturación')
sns.histplot(facturaciones, stat='probability')
```

```python
plt.title('distribución individual')
plt.xlabel('duración')
plt.xscale('log')
plt.yscale('log')
sns.histplot(duraciones)
```

Para estudiar una posible dependencia lineal entre las variables `facturacion` y `duracion`, usaremos dos herramientas. Primero realizaremos una estimación puntual de la pendiente que describe esta dependencia. En segundo lugar, la técnica de _remuestreo_ (_bootstrapping_) nos permitirá estudiar la distribución de la _pendiente_ de la recta que resulta al plantear un modelo de regresión lineal. Con esta última construiremos un intervalo de confianza para la pendiente.

```python
# Estimación puntual
lr = stats.linregress(duraciones, facturaciones)

lr.slope
```

```python
# Intervalo de confianza
import random

random_slopes = []
for i in range(10000):
    random_rows = random.choices(rows, k=len(rows))
    
    random_facturaciones = tuple(row[0] for row in random_rows)
    random_duraciones = tuple(row[1] for row in random_rows)

    lr = stats.linregress(random_duraciones, random_facturaciones)

    random_slopes.append(lr.slope)

sns.histplot(random_slopes)

sorted_random_slopes = sorted(random_slopes)
sorted_random_slopes[2500], sorted_random_slopes[len(sorted_random_slopes) - 2500]
```
