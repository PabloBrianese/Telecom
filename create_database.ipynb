Comenzamos inspeccionando los archivos `csv` usando los m√≥dulos `csv` y `pprint`, luego los transformamos en estructuras de datos de Python.

```python
import csv
from pprint import pprint

data = {}
```

```python
filename = "cliente.csv"
data["cliente"] = []

with open(filename) as file:
    rdr = csv.reader(file, delimiter=";")

    column_names = next(rdr)
    data["cliente"].append(column_names)
    
    rows = rdr
    data["cliente"] += [[int(row[0]), int(row[1]), int(row[2])] for row in rows]


pprint(data["cliente"])
```

```python
filename = "facturacion.csv"
data["facturacion"] = []

with open(filename) as file:
    rdr = csv.reader(file, delimiter=";")
    
    column_names = next(rdr)
    data["facturacion"].append(column_names)
    
    rows = rdr
    data["facturacion"] += [[float(row[0]), int(row[1]), int(row[2])] for row in rows]


pprint(data["facturacion"])
```

```python
filename = "llamadas.csv"
data["llamadas"] = []
with open(filename) as file:
    rdr = csv.reader(file, delimiter=";")
    
    column_names = next(rdr)
    data["llamadas"].append(column_names)
    
    rows = rdr
    data["llamadas"] += [[int(row[0]), int(row[1]), int(row[2]), int(row[3]), str(row[4])] for row in rows]

pprint(data["llamadas"])
```

```python
filename = "provincia.csv"
data["provincia"] = []
with open(filename) as file:
    rdr = csv.reader(file, delimiter=";")
    
    column_names = next(rdr)
    data["provincia"].append(column_names)
    
    rows = rdr
    data["provincia"] += [[str(row[0]), int(row[1])] for row in rows]

pprint(data["provincia"])
```

Teniendo los datos en un formato manipulable, procedemos a crear la base de datos. Es muy importante haber identificado los tipos de dato de las columnas para el correcto funcionamiento de la base de datos.

```python
def create_table(table_name, connection):
    column_names = data[table_name][0]
    rows = data[table_name][1:]

    create_table = "CREATE TABLE IF NOT EXISTS " + table_name + " (" + ",".join(column_names) + ");"
    insert_row = "INSERT INTO " + table_name + " VALUES (" + ",".join(len(column_names) * ["?"]) + ");"

    cur = connection.cursor()
    cur.execute(create_table)
    cur.executemany(insert_row, rows)

    connection.commit()


import sqlite3
with sqlite3.connect("telecom.db") as con:
    for table_name in data.keys(): create_table(table_name, con)
    
    pprint(con.execute("""
        SELECT * FROM sqlite_master;
    """).fetchall())
```
